```python
import os
from dotenv import load_dotenv
load_dotenv()

print("=" * 50)
print("ENVIRONMENT SETUP VERIFICATION")
print("=" * 50)

# 1. Core libraries
import pandas as pd
import numpy as np
import sklearn
import xgboost as xgb
import lightgbm as lgb
import shap
print(f" pandas         {pd.__version__}")
print(f" numpy          {np.__version__}")
print(f" scikit-learn   {sklearn.__version__}")
print(f" xgboost        {xgb.__version__}")
print(f" lightgbm       {lgb.__version__}")
print(f" shap           {shap.__version__}")

# 2. New GenAI libraries
import transformers
import sentence_transformers
import langchain
import praw
import fredapi
import groq
import scipy
print(f" transformers   {transformers.__version__}")
print(f" sentence-transformers {sentence_transformers.__version__}")
print(f" langchain      {langchain.__version__}")
print(f" praw           {praw.__version__}")
print(f" fredapi        {fredapi.__version__}")
print(f" groq           {groq.__version__}")
print(f" scipy          {scipy.__version__}")

# 3. Ollama
import requests
r = requests.post("http://localhost:11434/api/generate",
    json={"model": "llama3.2:3b", "prompt": "say ok", "stream": False})
print(f" Ollama         {r.json()['model']} — {r.json()['response'].strip()}")

# 4. FinBERT
from transformers import pipeline
pipe = pipeline("text-classification", model="ProsusAI/finbert")
result = pipe("The borrower may default on their loan.")
print(f" FinBERT        label={result[0]['label']} score={result[0]['score']:.3f}")

# 5. FRED
from fredapi import Fred
fred = Fred(api_key=os.getenv("FRED_API_KEY"))
s = fred.get_series("UNRATE", observation_start="2015-01-01", observation_end="2015-03-01")
print(f" FRED           UNRATE Jan-2015 = {s.iloc[0]}")

# 6. GDELT
import time
time.sleep(3)
r2 = requests.get(
    "https://api.gdeltproject.org/api/v2/doc/doc?query=credit+default&sourcelang=eng&mode=artlist&maxrecords=3&format=json",
    timeout=30)
if r2.status_code == 200 and r2.text.strip():
    articles = r2.json().get('articles', [])
    print(f" GDELT          status=200 articles={len(articles)}")
else:
    print(f" GDELT          status={r2.status_code} (rate limited — connectivity confirmed, skip for now)")

# 7. Groq
from groq import Groq
client = Groq(api_key=os.getenv("GROQ_API_KEY"))
resp = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[{"role": "user", "content": "say ok"}]
)
print(f" Groq           {resp.choices[0].message.content.strip()}")

# 8. Model artefacts
import pickle
for name, path in [
    ("lr_model",       "../models/lr_model.pkl"),
    ("xgb_model",      "../models/xgb_model.pkl"),
    ("shap_explainer", "../models/shap_explainer.pkl"),
    ("woe_mappings",   "../data/processed/woe_mappings.pkl"),
]:
    with open(path, "rb") as f:
        obj = pickle.load(f)
    print(f" {name:<16} loaded")

print()
print("=" * 50)
print("ALL CHECKS PASSED — READY TO BUILD")
print("=" * 50)
```
